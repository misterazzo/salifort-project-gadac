```python
# Create a boxplot to visualize distribution of `tenure` and detect any outliers
### YOUR CODE HERE ###
plt.figure(figsize=[5,1])
plt.title('Boxplot of tenure to detect its outliers')
sns.boxplot(x=df1['tenure'])
```

The **boxplot** displayed above shows that there are outliers for the variable *tenure*. Let's find out how many exactly with the code below.
NOTE: outliers in this example are considered as values with 1.5 IQR either above 75 quantile or below the 25 quantile.

```python
# Determine the number of rows containing outliers
### YOUR CODE HERE ###
percentile_25 = df1['tenure'].quantile(0.25)
percentile_75 = df1['tenure'].quantile(0.75)
iqr = percentile_75 - percentile_25

upper_limit = percentile_75 + 1.5*iqr
lower_limit = percentile_25 - 1.5*iqr
print('Lower limit: ', lower_limit)
print('Upper limit: ', upper_limit)

outliers = df1[(df1['tenure'] >= upper_limit) | (df1['tenure'] <= lower_limit)]
print('Number of outliers: ', len(outliers))
```

```python
# Create a plot as needed
### YOUR CODE HERE ###
sns.boxplot(data=df1, 
            x='satisfaction_level',
            y='salary',
            hue='left',
            orient='h')
plt.title('Distributions of satisfaction levels divided by salary for employees who left vs stayed')
```

All median and average values tend to stay constant for satisfaction levels no matter the salary between the group that leaves and the one that stays. Let's explore the impact of other variables such as how evenly distributed the people who leave are based on the department they work in.

```python
# Create a plot as needed
### YOUR CODE HERE ###
plt.figure(figsize=(15,7))
sns.histplot(data=df1, x='department', hue='left', multiple='dodge', shrink=0.5)
plt.title('Distribution of employees who left vs stayed by department')
```

It appears as if all departments have the same number of employees who stay and leave in percentage terms. Let's verify this quantitatively.

```python
left_ratio_dict = {}
for department in df1['department'].unique():
    department_subset = df1[df1['department'] == department]
    ratio = len(department_subset[department_subset['left'] == 1])/len(department_subset[department_subset['left'] == 0])
    left_ratio_dict[department] = np.round(ratio * 100.00, 2)
left_ratio_dict

```

From this calculation it appears that while most departments have about 1/5 of their employees leaving the company, R&D and management appear to be the ones with the lowest turnover. It is also true that since these are among the departments with less employees, the chance of seeing higher or lower than average values is greater.

## Recall evaluation metrics

- **AUC** is the area under the ROC curve; it's also considered the probability that the model ranks a random positive example more highly than a random negative example.
- **Precision** measures the proportion of data points predicted as True that are actually True, in other words, the proportion of positive predictions that are true positives.
- **Recall** measures the proportion of data points that are predicted as True, out of all the data points that are actually True. In other words, it measures the proportion of positives that are correctly classified.
- **Accuracy** measures the proportion of data points that are correctly classified.
- **F1-score** is an aggregation of precision and recall.
